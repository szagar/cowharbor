```python
def n_days(d1_str, d2_str):
    a = datetime.strptime(d1_str, zts.date_format)
    b = datetime.strptime(d2_str, zts.date_format)
    return abs((b - a).days)
```

```python

def add_days(dt_str, days):
    return (datetime.strptime(dt_str, zts.date_format) + timedelta(days=days)).strftime(
        zts.date_format
    )
```

```python
def opt_dates_from_oos_dates(oos_start_dt, oos_end_dt):
    tot_days = n_days(oos_start_dt, oos_end_dt)
    is_days = int(0.7 * tot_days)
    # oos_days = tot_days - is_days

    opt_start = oos_start_dt
    opt_end = add_days(opt_start, is_days)

    return (opt_start, opt_end)
```

```python




#def test_dates_from_opt_date(opt_end):
#    test_start = add_days(opt_end, 1)
#    test_end = (datetime.today() - timedelta(days=testing_days_ago)).strftime(
#        zts.date_format
#    )

#    return (test_start, test_end)


##
## data blocks
##
def initDataSets():
    data_sets = []
    data_sets.append([1, "1/1/08", "1/1/18"])
    data_sets.append([2, "1/1/08", "7/1/18"])
    data_sets.append([3, "7/1/08", "7/1/18"])
    data_sets.append([4, "7/1/08", "4/1/19"])
    data_sets.append([5, "7/1/08", "9/30/19"])
    df = pd.DataFrame(data_sets, columns=["data_set", "start_dt", "end_dt"])
    df.set_index("data_set", inplace=True)
    df["start_dt"] = pd.to_datetime(df["start_dt"])
    df["end_dt"] = pd.to_datetime(df["end_dt"])
    return df


def pick_a_block():
    return randint(1, 10)


def hhmm2mins(hhmm):
    i = int(hhmm)
    n_hrs = i // 100
    n_mins = i % 100
    return n_hrs * 60 + n_mins


def mins2hhmm(mins):
    hh = int(mins / 60)
    mm = mins - hh * 60
    return "%02d%02d" % (hh, mm)


def add2hhmm(hhmm, mins):
    return mins2hhmm(hhmm2mins(hhmm) + mins)


def lastDataBlock(data_sets):
    return max(data_sets.index)


def firstDataBlock():
    return min(data_sets.index)


def set_dates(data_sets, sess_st, sess_end, bars_back, data_set, data_block, entry_tf, use_daily):
    number_segments = 10
    date_format = "%m/%d/%Y"

    if use_daily:
        days_back = bars_back
    elif not use_daily:
        print(f"{sess_end} => {hhmm2mins(sess_end)}")
        print(f"{sess_st} => {hhmm2mins(sess_st)}")
        bars_per_session = int((hhmm2mins(sess_end) - hhmm2mins(sess_st)) / entry_tf)
        print(f"bars_per_session = {bars_per_session}")
        days_back = math.ceil(bars_back / bars_per_session)
        print(f"days_back = {days_back}")

    seg_size = int(
        (data_sets.loc[data_set]["end_dt"] - data_sets.loc[data_set]["start_dt"]).days
        / number_segments
    )
    start_dt = data_sets.loc[data_set]["start_dt"] + pd.DateOffset(
        (data_block - 1) * seg_size
    )
    pre_start_dt = start_dt - pd.DateOffset(round((days_back / 5) * 7))
    end_dt = start_dt + pd.DateOffset(seg_size)
    return {
        "pre_start_dt": pre_start_dt.strftime(date_format),
        "start_dt":     start_dt.strftime(date_format),
        "end_dt":       end_dt.strftime(date_format),
        "bt_start_dt":  data_sets.loc[data_set]["start_dt"].strftime(date_format),
        "bt_end_dt":    data_sets.loc[data_set]["end_dt"].strftime(date_format),
    }


```

```python
data_sets     = initDataSets()
session_start = 800
session_end   = 1300
max_bars_back = 200
data_set      = lastDataBlock(data_sets)
data_block    = pick_a_block()

entry_tf = 45
data_block    = pick_a_block()
print(set_dates(data_sets,session_start,session_end,max_bars_back,data_set,data_block,entry_tf,False))
data_block    = pick_a_block()
print(set_dates(data_sets,session_start,session_end,max_bars_back,data_set,data_block,entry_tf,False))

entry_tf = 45
print(set_dates(data_sets,session_start,session_end,max_bars_back,data_set,data_block,entry_tf,True))


session_start = 1000
session_end   = 1600
entry_tf = 15
print(set_dates(data_sets,session_start,session_end,max_bars_back,data_set,data_block,entry_tf,True))

print(pick_a_block())
print(pick_a_block())
```
